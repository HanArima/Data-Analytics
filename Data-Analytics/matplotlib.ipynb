# Matplotlib

- There are two ways to use the matplotlib library - Global API and OOP API

```python
import matplotlib.pyplot as plt
```

##### Basics


```python
x = [1,2,3,4,5,6,7,8]
y1 = [234,675,854,920,718,930,678,893]
y2 = [789, 678, 986,897,679,900,890,789]

plt.plot(x,y1, 'g-.', linewidth=4 ,label='Line1') # Format string parses the string in the following order color,marker,line
plt.plot(x,y2, 'c' ,label='Line2')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Sample Basic Plot')
```

#### GLOBAL API

```python

plt.figure(figsize=(20,5)) # Horizontal length, Vertical Length
plt.subplot(1,2,1) # rows, columns and panel selection
plt.subplot(1,2,2) # rows, columns and panel selection
```

- This is a figure and has 2 axes. In global API what happens is that we have to keep in mind what is happening from top down in each line. 
- We will have to refer to the subplots each time and tell the interpreter which figure are we refering to.  

#### OOPS API

- In this approach, we divide the figure in axes and save the reference to each axes for future reference to reference it later.

```python
fig, ((axes1, axes2), (axes3, axes4)) = plt.subplots(nrows=2, ncols=2, figsize=(20,10))

# Now we have created refernce for each of the subplot. BAsically variable names.
```

- figure() is a function that lets us set the aethetics of how our figure shall look like.
- In matplotlib, we have function for each graph.
