- **loc**: This lets you select the row by labels 
- **iloc**: This lets you select the row by specifying the integer location.

```python
(df.loc[df['Year'] == 2015])
```

### Drop Stuff

- A lot of times you might need to drop unnecessary rows and columns that are not that relevant.
- You cannot perform more than one drop_performance in one cell.
- Dropping a column doesn't change the df permanently until you save it as it is.

```python
df.drop(columns=['Day', 'Month', 'Year'])
```

```python

df.drop(columns=['Total_Cost'])
```

### Modifying DFs

- These changes are permanent and modifies the DF altogether.
- Like **drop, rename** is not a permanent change but changing the value is a permanent change on the df


df['Month'] = 'July'

```python
df['Calculated_Revenue'] = df['Order_Quantity'] * df['Unit_Price']
df
```
