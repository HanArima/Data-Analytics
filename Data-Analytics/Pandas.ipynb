# loc VS iloc
- **loc**: This lets you select the row by labels 
- **iloc**: This lets you select the row by specifying the integer location.

```python
df.head(2)
```

```python
(df.loc[df['Year'] == 2015])
```

```python
# Display all the females from US who bought Bikes in the Year 2015

df.loc[(df['Year'] == 2015 ) & (df['Customer_Gender'] == 'F') & (df['Country'] == 'United States') & (df['Product_Category'] == 'Bikes')]
```

```python
df.Product_Category.value_counts()
```

```python
df.Customer_Gender.value_counts()
```

```python
df.Product_Category.value_counts()
```

```python
# Display all the rows till 500

df.iloc[430:50]
```

```python
df.iloc[1]
```

```python
age_more_than_35 = df[df['Customer_Age'] > 35]
age_more_than_35
```
